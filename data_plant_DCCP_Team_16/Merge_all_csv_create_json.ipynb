# Importing libraries

```python
import pandas as pd
import glob
```

# Read all the csv files created by scraping and structed data and create combined data frame

```python
path = 'C:/Users/mrash/Desktop/ISB/DCCP/Group Project'
csv_files_list = glob.glob(path + "/*.csv")
df_list = (pd.read_csv(file) for file in csv_files_list)

combined_data   = pd.concat(df_list, ignore_index=True)
```

```python
combined_data.info()
```

# remove indexed column (Unnamed: 0)

```python
combined_data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

# data clean-up

```python
combined_data.fillna('', inplace=True)
combined_data["Medicinal_Use"].replace("None knownh3", "", regex=True , inplace=True)
combined_data["Medicinal_Use"].replace("None known", "", regex=True, inplace=True)
combined_data.drop_duplicates(keep='first' , inplace= True)
combined_data.drop_duplicates(subset=["Species_Name", "Family_name"], keep='first',  inplace= True)
```

# create a csv file for combine data after cleaning

```python
combined_data.to_csv('data_plant_combined_all.csv')
```

# Create json file for combine data after cleaning

```python
combined_data.to_json('data_plant_combined.json',orient ='records')
```
